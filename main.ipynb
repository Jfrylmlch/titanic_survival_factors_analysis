```python
from sklearn.datasets import fetch_openml
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
import numpy as np
import seaborn as sns
```

## Dataset Description

The Titanic dataset is a well-known collection of data detailing the passengers aboard the RMS Titanic, which tragically sank on its maiden voyage in 1912. This dataset is frequently used for data analysis and machine learning practice. Here's an explanation of each variable:

- **pclass**: Represents the passenger's ticket class, which is a proxy for socio-economic status. Values are:
  - 1 = 1st Class (Upper)
  - 2 = 2nd Class (Middle)
  - 3 = 3rd Class (Lower)

- **name**: The full name of the passenger.

- **sex**: The gender of the passenger. Values are typically 'male' or 'female'.

- **age**: The age of the passenger in years. Some entries may have fractional values for infants.

- **sibsp**: The number of siblings or spouses the passenger had aboard the Titanic.

- **parch**: The number of parents or children the passenger had aboard the Titanic.

- **ticket**: The ticket number assigned to the passenger.

- **fare**: The fare (ticket price) paid by the passenger.

- **cabin**: The cabin number where the passenger stayed. Many entries have missing or incomplete cabin information.

- **embarked**: The port where the passenger boarded the Titanic. Values are:
  - 'C' = Cherbourg
  - 'Q' = Queenstown
  - 'S' = Southampton

- **boat**: The lifeboat number the passenger boarded during the evacuation. Not all passengers were assigned lifeboats.

- **body**: An identification number assigned to the passenger's body after recovery. Many entries have missing body identification numbers.

- **home.dest**: The passenger's home or final destination. This field often contains concatenated strings of the passenger's home and destination locations.

## Introduction

This study aims to investigate the relationship between the backgrounds of Titanic passengers and their survival rates. By analyzing various demographic and social factors, we hope to uncover insights into how these elements influenced the likelihood of survival during the tragic sinking of the ship.

The dataset presents several challenges, particularly with missing values in key columns such as the body identification number, home or final destination, and cabin number. Due to the high volume of missing data in these areas, we will exclude them from our analysis. Additionally, the ticket number and fare are not central to our research questions and will also be disregarded.

To prepare the remaining data for analysis, we have implemented several cleaning steps:

- Rows lacking age information have been removed to ensure that our analysis includes only complete cases.
- The columns "sibsp" (the number of siblings or spouses aboard) and "parch" (the number of parents or children aboard) have been combined into a single variable, "relatives," which represents the total number of relatives a passenger had on board.
- People at different age were divided into age groups, i.e. 0-9, 10-19, 20-29...

Moreover, the availability of lifeboats is a critical factor influencing survival rates. While the dataset contains many missing values for the lifeboat numbers assigned to passengers, it is reasonable to assume that those without lifeboat information were not assigned to any lifeboat. Conversely, passengers with lifeboat information were likely assigned one, which will be factored into our analysis.

In summary, the cleaned dataset will focus on the following attributes:

- Sex
- Age Group
- Relatives
- Passenger Class (Pclass)

By concentrating on these variables, we aim to provide a clearer understanding of the factors that contributed to survival during the Titanic disaster.

```python
df = X.loc[:, ["pclass", "sex", "age", "parch", "sibsp", "boat"]]
df = pd.concat([df, y], axis=1)

df = df[~df["age"].isna()]

df["relatives"] = df["parch"] + df["sibsp"]
df.drop(["parch", "sibsp"], inplace=True, axis=1)

df["had_boat"] = ~df["boat"].isna()
df.drop("boat", inplace=True, axis=1)

df["age_group"] = df["age"] // 10 * 10
df["age_group"] = df["age_group"].map(lambda x: "%.0f-%.0f" % (x, x+9))
df.drop("age", axis=1, inplace=True)

df
```

## Effectiveness of lifeboats

By counting the number of survivors from the incident, it can be confirmed that passengers assigned to lifeboats had a high survival rate of 98%. In contrast, only 2.7% of passengers without lifeboat assignments were able to survive. This raises the question of how lifeboats were distributed during the abrupt onset of the disaster. By examining the factors related to the opportunity to access a lifeboat, the results may reflect some deep-rooted societal values of that period.

```python
df1 = df[["had_boat", "survived"]]
df1.loc[:, "survived"] = df1.loc[:, "survived"].astype(int)
df1 = pd.DataFrame(df1.value_counts().reset_index())
df1["survived"] = df1["survived"].map(lambda x: "survived" if x else "not survived")
df1["had_boat"] = df1["had_boat"].map(lambda x: "had boat" if x else "no boat")
df1 = pd.pivot_table(df1, values="count", index="had_boat", columns="survived")
sns.heatmap(df1, annot=True, fmt=".0f")

print("Survival rate given a lifeboat: %.2f%%" % (410/417 * 100))
print("Survival rate without a lifeboat: %.2f%%" % (17/629 * 100))
```

## Influence of passenger background on lifeboat access

The findings indicate that each factor significantly impacts access to lifeboats when considered individually.

Among the 1,046 passengers, 37.1% were female and 62.9% were male. Despite the smaller number of female passengers, they had nearly double the likelihood of accessing lifeboats compared to their male counterparts (26.4% vs. 13.5%). This suggests that, even amidst the chaos of the disaster, societal norms may have influenced the decision to prioritize women for lifeboat access, as females are often perceived as more vulnerable than males.

Another notable observation is that the more relatives a passenger had aboard the Titanic, the less likely they were to secure a lifeboat. In other words, a smaller family size appeared to increase the chances of obtaining a lifeboat. One possible explanation is that individuals with relatives aboard may have been preoccupied with the safety of their loved ones, potentially missing opportunities to board lifeboats. In contrast, single passengers without family ties may have been more flexible and less encumbered by concerns about family reunification. Furthermore, if families were considered as single units during the evacuation, rescuers may have prioritized the evacuation of families, thereby favoring those with fewer members.

Surprisingly, first-class passengers were the most likely to gain access to lifeboats, with 63% of them having this opportunity. They were the only class with a higher proportion of passengers accessing lifeboats, indicating that socio-economic status and wealth provided significant advantages in accessing emergency services. Interestingly, the likelihood of accessing a lifeboat did not decrease as passenger class lowered; in fact, third-class passengers had a higher chance of getting a lifeboat than those in second class. This phenomenon warrants further investigation in future research.

Regarding age, the findings revealed that young adults in the 20-29 age group were more likely to secure a lifeboat. This may be attributed to the relative agility and decisiveness of young adults, enabling them to navigate the chaotic environment effectively. Additionally, the limited number of lifeboats may have led older passengers and children to be more willing to sacrifice their chances of survival, as they might have perceived younger individuals as having a greater opportunity to survive.

```python
for c in ["pclass", "sex", "age_group", "relatives"]:
    temp = df[[c, "had_boat"]].value_counts().reset_index()
    hm = pd.pivot_table(temp, columns=c, index="had_boat", values="count", observed=False)
    fig, axes = plt.subplots(1, 2, layout="tight", figsize=(10, 5))
    sns.heatmap(hm, annot=True, fmt=".0f", ax=axes[1])
    display(hm)
    sns.barplot(temp[temp["had_boat"]==True], x=c, y="count", ax=axes[0], order=temp[c].sort_values())
    axes[0].set_ylabel("Number of people assigned lifeboats")
```
